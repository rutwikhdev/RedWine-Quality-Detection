## Inference

* The citric acid can add "freshness" and flavour to a wine and we saw that the quality of wine increases with the increase in citric acid content in wine.

* Similarly, the high value of volatile acid (acetic acid) can lead to unpleasant, vinegar taste and in our plot we observed that in high quality wine this acid content is low.

* Also, the pH value is low in good wines, in other words, the acidity of wine is high in good quality wines.

* In most cases, other values remains constant the increase in alcohol content increases the wine quality.

* And, sugar, salt, density all have a negative relationships to wine quality

* In the plots therse are too many outliers, as we know that the data is biased with more observations for normal wines.

```python
wine.quality.value_counts()
```

* We can see from the above counts that the quality rating 3, 4 and 8 has very less number of examples to learn.
* So I'm going to create two basic categories for quality (bad and good).
* If wine scores rating of 6 and above then it will have good quality. If less than 6 then bad rating.

```python
def set_quality(x):
    if x <= 5:
        return 0 # Bad
    else:
        return 1 # Good

wine['quality'] = wine['quality'].apply(set_quality)
```

```python
wine.quality.value_counts()
```

## 3. Model

### Dividing the dataset into train and test.

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(wine.drop(['quality'], axis=1), wine['quality'], 
                                       random_state=101, test_size=0.3)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=300, criterion='gini', random_state=41)
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
```

### Evaluating the model

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy')

print('Cross validation accuracy scores:',scores)
print('Test accuracy:', accuracy_score(y_test, preds))
```

### Sample predictions

```python
y_test[:5].values, preds[:5]
```

```python
from sklearn.metrics import f1_score

print('F1 score:',f1_score(y_test,preds, average='micro'))
```

```python
from sklearn.metrics import plot_confusion_matrix

plt.figure(figsize=(14,10))

disp = plot_confusion_matrix(clf, X_test, y_test, display_labels=['Bad', 'Good'], cmap=plt.cm.Blues)
disp.ax_.set_title('Confusion Matrix Red wine quality')

print(disp.confusion_matrix)

plt.show()
```

The model does about 81% accuracy on test set!
